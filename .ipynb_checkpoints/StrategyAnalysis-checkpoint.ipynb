```python
import os
import time
import json
from dataclasses import dataclass
from typing import Optional, Dict, Any

import requests
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfc = pd.read_csv("bot_trades_checkpoint.csv")
last_day = dfc[(dfc["timestamp"] < (1766176600+900)) & (dfc["timestamp"] > (1766176223 + 17 + 60 * 6))]
filtered_last_day = last_day[["timestamp", "size", "price", "side", "outcome", "slug"]]
filtered_last_day.head(10)
```

```python
print("min/max:", dfc["timestamp"].min(), dfc["timestamp"].max())
```

```python
filtered_last_day = filtered_last_day[
    filtered_last_day["slug"] == "btc-updown-15m-1766177100"
]
filtered_last_day.head(10)
```

```python
up_list = [0]
down_list = [0]
for index, row in filtered_last_day.iterrows():
    if row["outcome"] == "Up":
        up_list.append(up_list[-1] + row["size"])
        down_list.append(down_list[-1])
    else:
        down_list.append(down_list[-1] + row["size"])
        up_list.append(up_list[-1])
```

```python
filtered_last_day["up_volume"] = up_list[:-1]
filtered_last_day["down_volume"] = down_list[:-1]
filtered_last_day.head(10)
```

```python
len(filtered_last_day)
```

```python
df = filtered_last_day
plt.figure()
plt.plot(df["timestamp"], df["up_volume"], label="Up Volume")
plt.plot(df["timestamp"], df["down_volume"], label="Down Volume")

plt.xlabel("Timestamp")
plt.ylabel("Volume")
plt.legend()

plt.gca().invert_xaxis()
plt.tight_layout()
plt.show()
```

```python
reversed_15m = filtered_last_day.iloc[::-1]
reversed_15m.head(10)
```

```python
up_list = [0]
down_list = [0]
for index, row in reversed_15m.iterrows():
    if row["outcome"] == "Up":
        up_list.append(up_list[-1] + row["size"])
        down_list.append(down_list[-1])
    else:
        down_list.append(down_list[-1] + row["size"])
        up_list.append(up_list[-1])
```

```python
reversed_15m["up_volume"] = up_list[:-1]
reversed_15m["down_volume"] = down_list[:-1]
reversed_15m.head(10)
```

```python
df = reversed_15m
plt.figure()
plt.plot(df["timestamp"], df["up_volume"], label="Up Volume")
plt.plot(df["timestamp"], df["down_volume"], label="Down Volume")

plt.xlabel("Timestamp")
plt.ylabel("Volume")
plt.legend()

plt.tight_layout()
plt.show()
```
