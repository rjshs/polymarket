```python
import pandas as pd
import numpy as np
from pathlib import Path

# ============ CONFIGURATION ============
# Update these filenames to match your exact saved files
FILE_BTC = "polymarket_data/bitcoin_prices.parquet"
FILE_TRADES = "polymarket_data/trades_0x6031b6_1766362500_1766364300.parquet" # Update filename
FILE_POSITIONS = "polymarket_data/positions_0x6031b6_btc-updown-15m-17663.parquet" # Update filename
FILE_ORDERBOOK = "polymarket_data/orderbook_snapshots.parquet"

def load_and_standardize():
    print("ðŸ”„ Loading datasets...")

    # 1. LOAD TRADES (The Anchor)
    # This is our base table. We want to know market conditions for every single trade.
    df_trades = pd.read_parquet(FILE_TRADES)
    # Trades are likely in seconds, but let's ensure float
    df_trades['merge_time'] = df_trades['unix_timestamp'].astype(float)
    df_trades = df_trades.sort_values('merge_time')
    print(f"âœ… Trades Loaded: {len(df_trades)} rows")

    # 2. LOAD BTC PRICES
    df_btc = pd.read_parquet(FILE_BTC)
    # Fix: Convert Milliseconds to Seconds
    df_btc['merge_time'] = df_btc['server_timestamp'] / 1000.0
    df_btc = df_btc.sort_values('merge_time')
    # Keep only relevant columns to avoid clutter
    df_btc = df_btc[['merge_time', 'price']].rename(columns={'price': 'btc_oracle_price'})
    print(f"âœ… BTC Prices Loaded: {len(df_btc)} rows (Converted ms -> s)")

    # 3. LOAD ORDERBOOK
    df_ob = pd.read_parquet(FILE_ORDERBOOK)
    # It already has 'unixtime' in seconds
    df_ob['merge_time'] = df_ob['unixtime'].astype(float)
    df_ob = df_ob.sort_values('merge_time')
    print(f"âœ… Orderbook Loaded: {len(df_ob)} rows")

    # 4. LOAD POSITIONS
    df_pos = pd.read_parquet(FILE_POSITIONS)
    # Fix: Convert Datetime object to Unix Float
    # This handles the 'datetime.now()' vs UTC mismatch automatically
    df_pos['merge_time'] = df_pos['timestamp'].apply(lambda x: x.timestamp())
    df_pos = df_pos.sort_values('merge_time')
    print(f"âœ… Positions Loaded: {len(df_pos)} rows")

    return df_trades, df_btc, df_ob, df_pos

def merge_datasets(df_trades, df_btc, df_ob, df_pos):
    print("\nðŸ”— Merging datasets...")

    # A. Merge Trades + BTC Price
    # FIX: Tolerance is now 60.0 (seconds), because 'merge_time' is a float
    master = pd.merge_asof(
        df_trades,
        df_btc,
        on='merge_time',
        direction='backward',
        tolerance=60.0  # <--- CHANGED FROM pd.Timedelta
    )

    # B. Merge Trades + Orderbook
    master = pd.merge_asof(
        master,
        df_ob,
        on='merge_time',
        direction='backward',
        tolerance=5.0   # <--- CHANGED FROM pd.Timedelta
    )

    # C. Merge Trades + Positions
    master = pd.merge_asof(
        master,
        df_pos,
        on='merge_time',
        direction='backward',
        by='condition_id'
    )

    # CLEANUP: Calculate convenient metrics
    # 1. Price vs Oracle Delta
    master['arb_spread'] = master['price'] - master['btc_oracle_price']
    
    # 2. Maker vs Taker classification
    master['is_taker_buy'] = np.isclose(master['price'], master['up_best_ask'], atol=0.001)
    master['is_maker_buy'] = np.isclose(master['price'], master['up_best_bid'], atol=0.001)

    print(f"ðŸŽ‰ GRAND UNIFICATION COMPLETE. Master dataset size: {len(master)} rows")
    return master

# ============ EXECUTION ============
if __name__ == "__main__":
    # Load
    t, b, o, p = load_and_standardize()
    
    # Merge
    df_master = merge_datasets(t, b, o, p)
    
    # Preview
    cols_to_show = ['merge_time', 'side', 'size', 'price', 'btc_oracle_price', 'up_best_bid', 'up_best_ask', 'shares', 'cash_pnl']
    print("\nSample of Master Timeline:")
    print(df_master[cols_to_show].head(5))
    
    # Save for analysis
    df_master.to_parquet("polymarket_data/MASTER_ANALYSIS_DATASET.parquet")
    print("\nðŸ’¾ Saved to polymarket_data/MASTER_ANALYSIS_DATASET.parquet")
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

# ============ CONFIGURATION ============
DATA_DIR = Path("polymarket_data")
# Terms to search for (case-insensitive)
SEARCH_TERMS = ["eth", "ethereum"] 

def check_file_for_contamination(filepath):
    try:
        df = pd.read_parquet(filepath)
    except Exception as e:
        print(f"âš ï¸ Could not read {filepath.name}: {e}")
        return

    found_flags = []

    # 1. Check Column Names
    # We look for 'eth' in the column headers themselves
    for col in df.columns:
        for term in SEARCH_TERMS:
            if term in col.lower():
                found_flags.append(f"Column Name: '{col}'")

    # 2. Check Cell Values (String columns only)
    # We only scan columns that hold text (object/string) to save time
    str_cols = df.select_dtypes(include=['object', 'string']).columns
    
    for col in str_cols:
        # Check if any row in this column contains the term
        for term in SEARCH_TERMS:
            # case=False makes it case-insensitive (finds 'ETH', 'Ethereum', 'eth')
            matches = df[col].astype(str).str.contains(term, case=False, na=False)
            if matches.any():
                count = matches.sum()
                example = df.loc[matches, col].iloc[0]
                found_flags.append(f"Value in '{col}': found {count} rows (e.g., '{example}')")

    # 3. Report Results
    print(f"\nðŸ“„ Analyzing: {filepath.name}")
    if found_flags:
        print(f"   âŒ ETHEREUM DETECTED!")
        for flag in set(found_flags): # Use set to remove duplicates
            print(f"      - {flag}")
    else:
        print(f"   âœ… Clean (Bitcoin only)")

if __name__ == "__main__":
    print(f"ðŸ” Scanning {DATA_DIR} for Ethereum contamination...\n")
    
    # Grab all parquet files
    files = list(DATA_DIR.glob("*.parquet"))
    
    if not files:
        print(f"No parquet files found in {DATA_DIR}")
    else:
        for p_file in files:
            check_file_for_contamination(p_file)
            
    print("\nDone.")
```

```python
import pandas as pd
import numpy as np
import re
from datetime import datetime

# ---------------------------------------------------------
# 1. SETUP & LOAD
# ---------------------------------------------------------
input_file = 'polymarket_data/trades_0x6031b6_CLEANED_btc_15m_only.parquet'
output_file = 'polymarket_data/trades_0x6031b6_MASTER_ANALYSIS.parquet'

print("ðŸ”„ Loading data...")
df = pd.read_parquet(input_file)

# ---------------------------------------------------------
# 2. ROBUST COLUMN DETECTION
# ---------------------------------------------------------
# Find the column that contains the market names (e.g. 'market_slug', 'slug', 'token')
market_col = None
for col in df.columns:
    if df[col].dtype == 'object':
        # Check first non-null value for "btc" or "15m"
        val = df[col].dropna().iloc[0] if not df[col].dropna().empty else ""
        if isinstance(val, str) and "15m" in val:
            market_col = col
            break

if not market_col:
    raise KeyError(f"Could not identify market column. Available: {df.columns.tolist()}")

print(f"âœ… Target Market Column Identified: '{market_col}'")

# ---------------------------------------------------------
# 3. FEATURE ENGINEERING (The "Master Merge")
# ---------------------------------------------------------

# A. TIME HANDLING
# Ensure main timestamp is datetime
# (Adjust 'timestamp' below if your column is named 'date', 'created_at', etc.)
time_col = 'timestamp' if 'timestamp' in df.columns else df.select_dtypes(include=['datetime', 'int']).columns[0]
df['trade_time'] = pd.to_datetime(df[time_col], unit='s') if df[time_col].dtype != '<M8[ns]' else df[time_col]

# B. EXTRACT EXPIRY FROM SLUG
# The slug format is likely: "btc-updown-15m-1766362500"
# We extract the last 10 digits as the UNIX expiry time.
def extract_expiry(slug):
    match = re.search(r'(\d{10})', str(slug))
    return int(match.group(1)) if match else None

df['expiry_unix'] = df[market_col].apply(extract_expiry)
df['expiry_time'] = pd.to_datetime(df['expiry_unix'], unit='s')

# C. CALCULATE "TIME TO EXPIRY" (Crucial for 15m strat)
# How many seconds left in the market when the bot traded?
df['seconds_remaining'] = (df['expiry_time'] - df['trade_time']).dt.total_seconds()

# D. DIRECTIONAL EXPOSURE
# Standardize: Buying YES = Long (+1), Buying NO = Short (-1)
# Selling YES = Short (-1), Selling NO = Long (+1)
def get_signed_size(row):
    # Adjust column names 'side' and 'outcome' if necessary
    side = str(row.get('side', '')).upper()
    outcome = str(row.get('outcome', '')).upper()
    size = float(row.get('size', 0))
    
    if side == 'BUY':
        return size if outcome == 'YES' else -size
    elif side == 'SELL':
        return -size if outcome == 'YES' else size
    return 0

df['signed_size'] = df.apply(get_signed_size, axis=1)

# E. COST BASIS & VALUE
# 'price' * 'size' = volume (USDC committed)
df['trade_value_usdc'] = df['price'] * df['size']

# F. INVENTORY TRACKING (Running Position)
# We group by the specific market ID to see position accumulation per 15m round
df = df.sort_values(by=[market_col, 'trade_time'])
df['net_position_market'] = df.groupby(market_col)['signed_size'].cumsum()

# ---------------------------------------------------------
# 4. REVIEW & SAVE
# ---------------------------------------------------------
# Filter to columns we actually care about for strategy analysis
cols_to_keep = [
    'trade_time', market_col, 'expiry_time', 'seconds_remaining', 
    'side', 'outcome', 'price', 'size', 'signed_size', 
    'trade_value_usdc', 'net_position_market', 'transactionHash' # hash is useful for verifying unique txs
]

# Gracefully handle if some source cols are missing
final_cols = [c for c in cols_to_keep if c in df.columns]
df_final = df[final_cols].copy()

df_final.to_parquet(output_file)

print("-" * 50)
print(f"âœ… ANALYSIS COMPLETE. Saved to: {output_file}")
print("-" * 50)
print("First 5 rows of strategy data:")
print(df_final[['seconds_remaining', 'side', 'outcome', 'price', 'net_position_market']].head().to_string())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ---------------------------------------------------------
# 1. SETUP & LOAD
# ---------------------------------------------------------
input_file = 'polymarket_data/merged_analysis.parquet'
print(f"ðŸ”„ Loading {input_file}...")

try:
    df = pd.read_parquet(input_file)
except FileNotFoundError:
    print(f"âŒ File not found. Please ensure '{input_file}' exists.")
    exit()

# Ensure we have the right columns
if 'seconds_remaining' not in df.columns:
    print("âš ï¸ 'seconds_remaining' column missing. Recalculating...")
    # (Simplified recalculation if needed, assuming columns exist)
    # df['seconds_remaining'] = ... 
    pass 

# ---------------------------------------------------------
# 2. CALCULATE INTER-ARRIVAL TIMES
# ---------------------------------------------------------
# Sort by market and time to get correct deltas
df = df.sort_values(by=['slug', 'trade_time'])
df['prev_trade_time'] = df.groupby('slug')['trade_time'].shift(1)
df['time_delta'] = (df['trade_time'] - df['prev_trade_time']).dt.total_seconds()

# ---------------------------------------------------------
# 3. VISUALIZATION
# ---------------------------------------------------------
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# --- PLOT 1: THE HEARTBEAT (Frequency vs Time Remaining) ---
# We use 15-second bins (900s / 15 = 60 bins)
sns.histplot(df['seconds_remaining'], bins=60, color='#1f77b4', kde=True, ax=ax[0])
ax[0].set_xlim(900, 0)  # Invert X-axis: Start (900s) -> Expiry (0s)
ax[0].set_title('The "Heartbeat": When does the bot trade?', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Seconds Until Expiry (900s = Open, 0s = Close)')
ax[0].set_ylabel('Number of Trades')
ax[0].grid(axis='y', alpha=0.3)

# Add "Zone" Labels
ax[0].axvline(x=60, color='red', linestyle='--', alpha=0.5)
ax[0].text(30, ax[0].get_ylim()[1]*0.9, 'KILL ZONE\n(Last 60s)', color='red', ha='center', fontweight='bold')
ax[0].text(850, ax[0].get_ylim()[1]*0.9, 'OPENING\nLIQUIDITY', color='green', ha='center', fontweight='bold')

# --- PLOT 2: THE SPEED TEST (Inter-arrival Times) ---
# Log scale often works best for time deltas
sns.histplot(df['time_delta'].dropna(), bins=100, color='#ff7f0e', log_scale=(False, True), ax=ax[1])
ax[1].set_title('Infrastructure Speed: Time between trades', fontsize=14, fontweight='bold')
ax[1].set_xlabel('Seconds between consecutive trades')
ax[1].set_ylabel('Count (Log Scale)')
ax[1].set_xlim(0, 60) # Focus on 0-60 seconds gaps

plt.tight_layout()
plt.savefig('phase1_temporal_structure.png')
print("âœ… Graph saved: phase1_temporal_structure.png")

# ---------------------------------------------------------
# 4. STATISTICAL SUMMARY
# ---------------------------------------------------------
total_trades = len(df)
last_minute = len(df[df['seconds_remaining'] < 60])
first_minute = len(df[df['seconds_remaining'] > 840])
median_delta = df['time_delta'].median()
fast_trades = len(df[df['time_delta'] < 1.0])

print("\n" + "="*40)
print("ðŸ“Š TEMPORAL PROFILE REPORT")
print("="*40)
print(f"Total Trades Analyzed: {total_trades}")
print("-" * 30)
print(f"1. TRIGGER REGIME:")
print(f"   - Opening Rush (>840s): {first_minute} trades ({first_minute/total_trades:.1%})")
print(f"   - The Kill Zone (<60s): {last_minute} trades ({last_minute/total_trades:.1%})")
print("-" * 30)
print(f"2. INFRASTRUCTURE SPEED:")
print(f"   - Median Time Between Trades: {median_delta:.2f} seconds")
print(f"   - HFT Bursts (<1s gaps):      {fast_trades} times")
print("="*40)
```
