```python
import json
import threading
import pandas as pd
import time
from datetime import datetime
from pathlib import Path
from websocket import WebSocketApp

# ============ CONFIGURATION ============
TOKEN_ID_UP = "97253427119455342162871389373778666616857779934886935013565914617734970049311"
TOKEN_ID_DOWN = "32667348465546219279476150869101237294147392011954036789055714980648226426729"
MARKET_SLUG = "btc-updown-15m-1766475900"

DATA_DIR = Path("polymarket_data")
DATA_DIR.mkdir(exist_ok=True)
FILENAME = DATA_DIR / "orderbook_snapshots_new.parquet"

# ============ SHARED STATE ============
# Store best bid/ask from price_change events (CORRECT data)
prices = {
    TOKEN_ID_UP: {"best_bid": None, "best_ask": None},
    TOKEN_ID_DOWN: {"best_bid": None, "best_ask": None},
}
# Store full book for depth (may be stale, use with caution)
books = {
    TOKEN_ID_UP: {"bids": [], "asks": []},
    TOKEN_ID_DOWN: {"bids": [], "asks": []},
}
data_lock = threading.Lock()

snapshot_buffer = []
last_save_time = time.time()

# ============ WEBSOCKET ============
def on_open(ws):
    print("[WS] Connected")
    msg = {
        "assets_ids": [TOKEN_ID_UP, TOKEN_ID_DOWN],
        "type": "market"
    }
    ws.send(json.dumps(msg))
    
    # Start ping thread to keep connection alive
    def ping():
        while True:
            try:
                ws.send("PING")
                time.sleep(10)
            except:
                break
    threading.Thread(target=ping, daemon=True).start()

def on_message(ws, message):
    if message == "PONG":
        return

    try:
        data = json.loads(message)
        events = data if isinstance(data, list) else [data]

        for event in events:
            event_type = event.get("event_type")
            
            # Handle price_change events - THIS HAS CORRECT PRICES
            if event_type == "price_change":
                for change in event.get("price_changes", []):
                    asset_id = change.get("asset_id")
                    if asset_id in prices:
                        with data_lock:
                            if change.get("best_bid"):
                                prices[asset_id]["best_bid"] = float(change["best_bid"])
                            if change.get("best_ask"):
                                prices[asset_id]["best_ask"] = float(change["best_ask"])
                        print(f"[PRICE] {asset_id[:20]}... bid={change.get('best_bid')} ask={change.get('best_ask')}")
            
            # Handle last_trade_price events - also useful
            elif event_type == "last_trade_price":
                asset_id = event.get("asset_id")
                price = event.get("price")
                print(f"[TRADE] {asset_id[:20]}... price={price} size={event.get('size')}")
            
            # Handle book events - WARNING: often stale!
            elif event_type == "book":
                asset_id = event.get("asset_id")
                if asset_id in books:
                    with data_lock:
                        books[asset_id]["bids"] = event.get("bids") or []
                        books[asset_id]["asks"] = event.get("asks") or []
                    # Log warning about potentially stale data
                    bids = event.get("bids", [])
                    asks = event.get("asks", [])
                    if bids and asks:
                        best_bid = float(bids[0]["price"]) if bids else 0
                        best_ask = float(asks[0]["price"]) if asks else 0
                        # Detect ghost market (0.01/0.99 spread = stale)
                        if best_ask - best_bid > 0.9:
                            print(f"[BOOK WARNING] Stale data detected! bid={best_bid} ask={best_ask}")
                        else:
                            print(f"[BOOK] {asset_id[:20]}... bid={best_bid} ask={best_ask}")

    except Exception as e:
        print(f"[WS ERROR] {e}")

def run_websocket():
    while True:
        try:
            ws = WebSocketApp(
                "wss://ws-subscriptions-clob.polymarket.com/ws/market",
                on_open=on_open,
                on_message=on_message
            )
            ws.run_forever()
            print("[WS] Reconnecting in 2s...")
            time.sleep(2)
        except Exception as e:
            print(f"[WS CRASH] {e}")
            time.sleep(5)

# ============ SAMPLER & SAVER ============
def sampler_and_saver():
    global snapshot_buffer, last_save_time
    print("[SAMPLER] Started...")

    while True:
        now_ts = time.time()

        with data_lock:
            up_bid = prices[TOKEN_ID_UP]["best_bid"]
            up_ask = prices[TOKEN_ID_UP]["best_ask"]
            down_bid = prices[TOKEN_ID_DOWN]["best_bid"]
            down_ask = prices[TOKEN_ID_DOWN]["best_ask"]

        # Only record if we have price data from price_change events
        if up_bid is not None and up_ask is not None:
            record = {
                "timestamp": datetime.fromtimestamp(now_ts),
                "unixtime": now_ts,
                "market_slug": MARKET_SLUG,
                
                # Up Token (from price_change - CORRECT)
                "up_best_bid": up_bid,
                "up_best_ask": up_ask,
                "up_spread": up_ask - up_bid,
                
                # Down Token (from price_change - CORRECT)
                "down_best_bid": down_bid,
                "down_best_ask": down_ask,
                "down_spread": (down_ask - down_bid) if down_bid and down_ask else None,
                
                # Implied probability check
                "implied_sum_bid": up_bid + (down_bid or 0),
                "implied_sum_ask": up_ask + (down_ask or 0),
            }
            snapshot_buffer.append(record)

        # Save to disk
        if len(snapshot_buffer) > 0 and (now_ts - last_save_time > 60 or len(snapshot_buffer) >= 1000):
            print(f" Flushing {len(snapshot_buffer)} snapshots...")
            try:
                new_df = pd.DataFrame(snapshot_buffer)
                if FILENAME.exists():
                    existing_df = pd.read_parquet(FILENAME)
                    final_df = pd.concat([existing_df, new_df], ignore_index=True)
                else:
                    final_df = new_df
                final_df.to_parquet(FILENAME, index=False)
                print(" Saved.")
                snapshot_buffer = []
                last_save_time = now_ts
            except Exception as e:
                print(f" Save Failed: {e}")

        time.sleep(0.1)

# ============ EXECUTION ============
if __name__ == "__main__":
    t_ws = threading.Thread(target=run_websocket, daemon=True)
    t_ws.start()
    
    try:
        sampler_and_saver()
    except KeyboardInterrupt:
        print("Stopping...")
        if snapshot_buffer:
            print("Saving final data...")
            pd.DataFrame(snapshot_buffer).to_parquet(FILENAME, index=False)
```
