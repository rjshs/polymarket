```python
import pandas as pd
import matplotlib.pyplot as plt

prices = pd.read_parquet("DataCollection/BitcoinPrices/1766688300-prices.parquet")
trades = pd.read_parquet("DataCollection/BotTrades/1766688300-trades.parquet")

# Merge to get BTC price at each trade time
trades_sorted = trades.sort_values('unix_timestamp')
trades_sorted['unix_timestamp_ms'] = trades_sorted['unix_timestamp'] * 1000

prices_sorted = prices.sort_values('server_timestamp')

merged = pd.merge_asof(trades_sorted, prices_sorted, 
                        left_on='unix_timestamp_ms', right_on='server_timestamp')

# Plot using the millisecond timestamp
plt.figure(figsize=(14, 6))
plt.plot(prices['server_timestamp'], prices['price'], color='gray', alpha=0.7)

up = merged[merged['outcome'] == 'Up']
down = merged[merged['outcome'] == 'Down']

plt.scatter(up['unix_timestamp_ms'], up['price_y'], color='green', marker='^', s=50, label='Up', alpha=0.7)
plt.scatter(down['unix_timestamp_ms'], down['price_y'], color='red', marker='v', s=50, label='Down', alpha=0.7)

# specify start and end time
start_sec = 1766688400
end_sec = 1766688500
plt.xlim(start_sec * 1000, end_sec * 1000)

# plot
plt.legend()
plt.xlabel('Time')
plt.ylabel('BTC Price')
plt.title('BTC Price with Bot Trades')
plt.tight_layout()
plt.show()

# Look at trades at the same timestamp
paired = merged.groupby('unix_timestamp').agg({
    'outcome': list,
    'size': list,
    'price_x': list  # trade price, not BTC price
})
print(paired[paired['outcome'].apply(lambda x: 'Up' in x and 'Down' in x)].head(10))

# Calculate $ spent on each side per timestamp
def calc_exposure(row):
    up_spent = sum(s * p for s, p, o in zip(row['size'], row['price_x'], row['outcome']) if o == 'Up')
    down_spent = sum(s * p for s, p, o in zip(row['size'], row['price_x'], row['outcome']) if o == 'Down')
    return pd.Series({'up_$': up_spent, 'down_$': down_spent, 'ratio': up_spent / down_spent if down_spent > 0 else None})

exposure = paired.apply(calc_exposure, axis=1)
print(exposure.head(10))
print(f"\nTotal Up $: {exposure['up_$'].sum():.2f}")
print(f"Total Down $: {exposure['down_$'].sum():.2f}")
```

```python
import pandas as pd

trades = pd.read_parquet("DataCollection/BotTrades/1766688300-trades.parquet")
orderbook = pd.read_parquet("DataCollection/OrderbookData/1766688300-orderbook.parquet")

# Merge trades with orderbook state
orderbook['unix_ts'] = orderbook['unixtime'].astype(int)
trades_sorted = trades.sort_values('unix_timestamp')
orderbook_sorted = orderbook.sort_values('unix_ts')

merged = pd.merge_asof(trades_sorted, orderbook_sorted,
                        left_on='unix_timestamp', right_on='unix_ts',
                        direction='backward')

# For each trade, calculate: did they pay the ask (taker) or better (maker)?
def classify_trade(row):
    if row['outcome'] == 'Up':
        if row['price'] >= row['up_best_ask']:
            return 'taker'
        elif row['price'] <= row['up_best_bid']:
            return 'maker_filled'  # their bid got hit
        else:
            return 'mid'  # traded inside spread
    else:  # Down
        if row['price'] >= row['down_best_ask']:
            return 'taker'
        elif row['price'] <= row['down_best_bid']:
            return 'maker_filled'
        else:
            return 'mid'

merged['order_type'] = merged.apply(classify_trade, axis=1)
print("=== ORDER TYPE DISTRIBUTION ===")
print(merged['order_type'].value_counts(normalize=True))

# Does size correlate with spread opportunity?
merged['combined_spread'] = merged['implied_sum_ask'] - merged['implied_sum_bid']
print("\n=== SIZE vs SPREAD CORRELATION ===")
print(f"Correlation: {merged['size'].corr(merged['combined_spread']):.3f}")

# Size by order type - do they size differently for taker vs maker?
print("\n=== SIZE BY ORDER TYPE ===")
print(merged.groupby('order_type')['size'].describe())

# Is 28 a hard cap?
print(f"\n=== SIZE CAP ===")
print(f"Max size: {merged['size'].max()}")
print(f"Trades at max: {(merged['size'] == 28).sum()} / {len(merged)}")

print(trades['size'].value_counts().head(20))
```

```python
import pandas as pd
import numpy as np

# Load all markets
markets = ["1766688300", "1766689200", "1766690100", "1766691000", "1766691900", "1766692800"]
all_trades = []
all_merged = []

for market in markets:
    try:
        trades = pd.read_parquet(f"DataCollection/BotTrades/{market}-trades.parquet")
        orderbook = pd.read_parquet(f"DataCollection/OrderbookData/{market}-orderbook.parquet")
        
        trades['market'] = market
        
        # Merge with orderbook
        orderbook['unix_ts'] = orderbook['unixtime'].astype(int)
        trades_sorted = trades.sort_values('unix_timestamp')
        orderbook_sorted = orderbook.sort_values('unix_ts')
        
        merged = pd.merge_asof(trades_sorted, orderbook_sorted,
                                left_on='unix_timestamp', right_on='unix_ts',
                                direction='backward')
        merged['market'] = market
        
        all_trades.append(trades)
        all_merged.append(merged)
    except:
        pass

trades = pd.concat(all_trades, ignore_index=True)
merged = pd.concat(all_merged, ignore_index=True)

print(f"Total trades across all markets: {len(trades)}")

# ============================================
# PART 1: SIZE PATTERN IDENTIFICATION
# ============================================
print("\n" + "="*60)
print("PART 1: SIZE PATTERNS")
print("="*60)

# Separate round vs decimal sizes
def is_round_size(size):
    return size == round(size, 0) or size in [1.21]  # 1.21 appears intentional

merged['is_round'] = merged['size'].apply(lambda x: x == round(x, 0))
merged['size_rounded'] = merged['size'].round(0)

print("\n=== ROUND vs FRACTIONAL ===")
print(merged['is_round'].value_counts(normalize=True))

print("\n=== TOP 20 EXACT SIZES (all markets) ===")
print(trades['size'].value_counts().head(20))

# Check if fractional sizes are clustered near round numbers (partial fills)
fractional = merged[~merged['is_round']]
print(f"\n=== FRACTIONAL SIZE ANALYSIS ===")
print(f"Count: {len(fractional)}")
print(f"Mean distance from nearest integer: {(fractional['size'] - fractional['size'].round()).abs().mean():.3f}")

# ============================================
# PART 2: CONSECUTIVE TRADE GROUPING
# ============================================
print("\n" + "="*60)
print("PART 2: CONSECUTIVE TRADE RECONSTRUCTION")
print("="*60)

# Group trades by timestamp to see original order intentions
trades_by_time = merged.groupby(['market', 'unix_timestamp', 'outcome']).agg({
    'size': ['sum', 'count', 'mean', 'max'],
    'price': 'mean'
}).reset_index()
trades_by_time.columns = ['market', 'timestamp', 'outcome', 'total_size', 'num_fills', 'avg_fill', 'max_fill', 'avg_price']

print("\n=== TRADES GROUPED BY TIMESTAMP ===")
print(f"Total grouped orders: {len(trades_by_time)}")
print(f"Single fills: {(trades_by_time['num_fills'] == 1).sum()}")
print(f"Multiple fills: {(trades_by_time['num_fills'] > 1).sum()}")

print("\n=== GROUPED ORDER SIZE DISTRIBUTION ===")
print(trades_by_time['total_size'].describe(percentiles=[.25, .5, .75, .9, .95, .99]))

print("\n=== NUM FILLS PER ORDER ===")
print(trades_by_time['num_fills'].value_counts().head(10))

# Does grouped size cluster at 28?
print("\n=== GROUPED SIZES NEAR 28 ===")
print(f"Grouped size == 28: {(trades_by_time['total_size'] == 28).sum()}")
print(f"Grouped size 27-29: {((trades_by_time['total_size'] >= 27) & (trades_by_time['total_size'] <= 29)).sum()}")
print(f"Grouped size > 28: {(trades_by_time['total_size'] > 28).sum()}")

# ============================================
# PART 3: POSITION-BASED SIZING
# ============================================
print("\n" + "="*60)
print("PART 3: POSITION-BASED SIZING (REBALANCING)")
print("="*60)

# Calculate running position at each trade
for market_id in merged['market'].unique():
    mask = merged['market'] == market_id
    market_trades = merged[mask].sort_values('unix_timestamp')
    
    up_cumsum = (market_trades['outcome'] == 'Up').astype(int) * market_trades['size']
    down_cumsum = (market_trades['outcome'] == 'Down').astype(int) * market_trades['size']
    
    merged.loc[mask, 'cumulative_up'] = up_cumsum.cumsum()
    merged.loc[mask, 'cumulative_down'] = down_cumsum.cumsum()

merged['net_exposure'] = merged['cumulative_up'] - merged['cumulative_down']
merged['abs_exposure'] = merged['net_exposure'].abs()

# Does size correlate with exposure (bigger trades when imbalanced)?
print(f"\n=== SIZE vs EXPOSURE CORRELATION ===")
print(f"Size vs Net exposure: {merged['size'].corr(merged['net_exposure']):.3f}")
print(f"Size vs Abs exposure: {merged['size'].corr(merged['abs_exposure']):.3f}")

# Do they size UP on the underweight side?
merged['is_up_trade'] = merged['outcome'] == 'Up'
merged['exposure_before'] = merged['net_exposure'] - merged.apply(
    lambda x: x['size'] if x['outcome'] == 'Up' else -x['size'], axis=1)

# When net exposure is negative (underweight Up), are Up trades bigger?
underweight_up = merged[merged['exposure_before'] < -50]
underweight_down = merged[merged['exposure_before'] > 50]

print(f"\n=== REBALANCING BEHAVIOR ===")
print(f"When underweight Up (exposure < -50):")
if len(underweight_up) > 0:
    print(f"  Up trade avg size: {underweight_up[underweight_up['outcome']=='Up']['size'].mean():.2f}")
    print(f"  Down trade avg size: {underweight_up[underweight_up['outcome']=='Down']['size'].mean():.2f}")

print(f"\nWhen underweight Down (exposure > 50):")
if len(underweight_down) > 0:
    print(f"  Up trade avg size: {underweight_down[underweight_down['outcome']=='Up']['size'].mean():.2f}")
    print(f"  Down trade avg size: {underweight_down[underweight_down['outcome']=='Down']['size'].mean():.2f}")

# ============================================
# PART 4: TIMING-BASED SIZING
# ============================================
print("\n" + "="*60)
print("PART 4: TIMING PATTERNS")
print("="*60)

# Normalize timestamp within each market window (0 = start, 1 = end)
for market_id in merged['market'].unique():
    mask = merged['market'] == market_id
    min_ts = merged.loc[mask, 'unix_timestamp'].min()
    max_ts = merged.loc[mask, 'unix_timestamp'].max()
    merged.loc[mask, 'time_pct'] = (merged.loc[mask, 'unix_timestamp'] - min_ts) / (max_ts - min_ts + 1)

# Size by time bucket
merged['time_bucket'] = pd.cut(merged['time_pct'], bins=[0, 0.25, 0.5, 0.75, 1.0], labels=['Q1', 'Q2', 'Q3', 'Q4'])

print("\n=== SIZE BY TIME QUARTILE ===")
print(merged.groupby('time_bucket')['size'].agg(['mean', 'std', 'count']))

print(f"\n=== SIZE vs TIME CORRELATION ===")
print(f"Size vs Time %: {merged['size'].corr(merged['time_pct']):.3f}")

# ============================================
# PART 5: ORDERBOOK-BASED SIZING
# ============================================
print("\n" + "="*60)
print("PART 5: ORDERBOOK RELATIONSHIP")
print("="*60)

# Calculate spread at trade time
merged['up_spread'] = merged['up_best_ask'] - merged['up_best_bid']
merged['down_spread'] = merged['down_best_ask'] - merged['down_best_bid']
merged['arb_margin'] = 1 - merged['implied_sum_ask']

print(f"\n=== SIZE vs ORDERBOOK CORRELATIONS ===")
print(f"Size vs Up spread: {merged['size'].corr(merged['up_spread']):.3f}")
print(f"Size vs Down spread: {merged['size'].corr(merged['down_spread']):.3f}")
print(f"Size vs Arb margin: {merged['size'].corr(merged['arb_margin']):.3f}")

# Size when arb is available vs not
arb_available = merged[merged['implied_sum_ask'] < 1.0]
no_arb = merged[merged['implied_sum_ask'] >= 1.0]

print(f"\n=== SIZE WHEN ARB AVAILABLE vs NOT ===")
print(f"Arb available (sum < 1.0): avg size = {arb_available['size'].mean():.2f}, count = {len(arb_available)}")
print(f"No arb (sum >= 1.0): avg size = {no_arb['size'].mean():.2f}, count = {len(no_arb)}")

# ============================================
# PART 6: TAKER vs MAKER SIZING
# ============================================
print("\n" + "="*60)
print("PART 6: EXECUTION TYPE SIZING")
print("="*60)

def classify_trade(row):
    if row['outcome'] == 'Up':
        if row['price'] >= row['up_best_ask']:
            return 'taker'
        elif row['price'] <= row['up_best_bid']:
            return 'maker'
        else:
            return 'mid'
    else:
        if row['price'] >= row['down_best_ask']:
            return 'taker'
        elif row['price'] <= row['down_best_bid']:
            return 'maker'
        else:
            return 'mid'

merged['exec_type'] = merged.apply(classify_trade, axis=1)

print("\n=== SIZE BY EXECUTION TYPE ===")
print(merged.groupby('exec_type')['size'].agg(['mean', 'std', 'count']))

# Are max-size trades more likely to be taker or maker?
max_size_trades = merged[merged['size'] == 28]
print(f"\n=== MAX SIZE (28) TRADES EXECUTION TYPE ===")
print(max_size_trades['exec_type'].value_counts(normalize=True))

# ============================================
# SUMMARY
# ============================================
print("\n" + "="*60)
print("SUMMARY: SIZING STRATEGY CONCLUSIONS")
print("="*60)
```

```python
"""
TEST 1: QUOTE PLACEMENT LOGIC
=============================
Question: Where exactly do they post relative to best bid/ask?

Method: Compare their fill prices to orderbook state at time of fill.
- If fill_price == best_bid  posting AT the bid
- If fill_price > best_bid  posting INSIDE the spread
- If fill_price < best_bid  getting filled at worse price (unusual)
"""

import pandas as pd
import numpy as np

markets = ["1766688300", "1766689200", "1766690100", "1766691000", "1766691900", "1766692800"]

results = []

for market in markets:
    try:
        trades = pd.read_parquet(f"DataCollection/BotTrades/{market}-trades.parquet")
        orderbook = pd.read_parquet(f"DataCollection/OrderbookData/{market}-orderbook.parquet")
        
        # Align trades with orderbook state
        trades = trades.sort_values('unix_timestamp')
        orderbook = orderbook.sort_values('unixtime')
        orderbook['unix_ts'] = orderbook['unixtime'].astype(int)
        
        merged = pd.merge_asof(
            trades,
            orderbook[['unix_ts', 'up_best_bid', 'up_best_ask', 'down_best_bid', 'down_best_ask']],
            left_on='unix_timestamp',
            right_on='unix_ts',
            direction='backward'
        )
        
        for _, row in merged.iterrows():
            if row['outcome'] == 'Up':
                best_bid = row['up_best_bid']
                best_ask = row['up_best_ask']
            else:
                best_bid = row['down_best_bid']
                best_ask = row['down_best_ask']
            
            fill_price = row['price']
            spread = best_ask - best_bid if pd.notna(best_ask) and pd.notna(best_bid) else None
            
            if pd.notna(best_bid) and pd.notna(best_ask):
                # Where in the spread did they get filled?
                if spread > 0:
                    spread_position = (fill_price - best_bid) / spread  # 0 = at bid, 1 = at ask
                else:
                    spread_position = None
                    
                distance_from_bid = fill_price - best_bid
                distance_from_ask = fill_price - best_ask
                
                results.append({
                    'market': market,
                    'outcome': row['outcome'],
                    'fill_price': fill_price,
                    'best_bid': best_bid,
                    'best_ask': best_ask,
                    'spread': spread,
                    'distance_from_bid': distance_from_bid,
                    'distance_from_ask': distance_from_ask,
                    'spread_position': spread_position
                })
    except Exception as e:
        print(f"Error processing {market}: {e}")

df = pd.DataFrame(results)

print("="*70)
print("TEST 1: QUOTE PLACEMENT ANALYSIS")
print("="*70)

print("\n=== DISTANCE FROM BEST BID (where they got filled vs visible bid) ===")
print(df['distance_from_bid'].describe())

print("\n=== BREAKDOWN BY POSITION ===")
at_bid = (df['distance_from_bid'].abs() < 0.001).sum()
inside_spread = ((df['distance_from_bid'] > 0.001) & (df['distance_from_ask'] < -0.001)).sum()
at_ask = (df['distance_from_ask'].abs() < 0.001).sum()
above_ask = (df['distance_from_ask'] > 0.001).sum()
below_bid = (df['distance_from_bid'] < -0.001).sum()

total = len(df)
print(f"Filled AT best bid (0.001):     {at_bid:>5} ({at_bid/total*100:>5.1f}%)")
print(f"Filled INSIDE spread:            {inside_spread:>5} ({inside_spread/total*100:>5.1f}%)")
print(f"Filled AT best ask:              {at_ask:>5} ({at_ask/total*100:>5.1f}%)")
print(f"Filled ABOVE ask (crossed):      {above_ask:>5} ({above_ask/total*100:>5.1f}%)")
print(f"Filled BELOW bid (got better):   {below_bid:>5} ({below_bid/total*100:>5.1f}%)")

print("\n=== SPREAD POSITION (0=bid, 0.5=mid, 1=ask) ===")
valid_spread_pos = df[df['spread_position'].notna()]['spread_position']
print(valid_spread_pos.describe())

print("\n=== BY OUTCOME ===")
for outcome in ['Up', 'Down']:
    subset = df[df['outcome'] == outcome]
    print(f"\n{outcome}:")
    print(f"  Avg distance from bid: {subset['distance_from_bid'].mean():.4f}")
    print(f"  Avg spread position:   {subset[subset['spread_position'].notna()]['spread_position'].mean():.3f}")

print("\n" + "="*70)
print("INTERPRETATION:")
print("="*70)
print("""
- If most fills are AT bid (distance  0): They post at best bid
- If most fills are INSIDE spread: They post aggressively inside spread
- If most fills are BELOW bid: Orderbook data is lagged (they're getting better prices than we see)
- Spread position near 0.5: They're posting at midpoint
""")
```

```python
"""
TEST 2: ORDER REFRESH / REPOST TIMING
=====================================
Question: How fast do they repost after a fill?

Method: 
- Group consecutive fills on the same side
- Measure time gaps between fills
- Infer repost speed from fill patterns
"""

import pandas as pd
import numpy as np

markets = ["1766688300", "1766689200", "1766690100", "1766691000", "1766691900", "1766692800"]

all_trades = []
for market in markets:
    try:
        trades = pd.read_parquet(f"DataCollection/BotTrades/{market}-trades.parquet")
        trades['market'] = market
        all_trades.append(trades)
    except:
        pass

trades = pd.concat(all_trades, ignore_index=True)
trades = trades.sort_values('unix_timestamp')

print("="*70)
print("TEST 2: ORDER REFRESH TIMING")
print("="*70)

# Time between consecutive fills on SAME side
for outcome in ['Up', 'Down']:
    subset = trades[trades['outcome'] == outcome].copy()
    subset = subset.sort_values('unix_timestamp')
    subset['time_since_last'] = subset['unix_timestamp'].diff()
    
    print(f"\n=== {outcome} - Time Between Consecutive Fills (seconds) ===")
    print(subset['time_since_last'].describe())
    
    print(f"\nDistribution:")
    bins = [0, 0.5, 1, 2, 5, 10, 30, 60, float('inf')]
    labels = ['<0.5s', '0.5-1s', '1-2s', '2-5s', '5-10s', '10-30s', '30-60s', '>60s']
    subset['time_bucket'] = pd.cut(subset['time_since_last'], bins=bins, labels=labels)
    print(subset['time_bucket'].value_counts().sort_index())

# Now look at ALTERNATING pattern (Up then Down, etc.)
print("\n" + "="*70)
print("ALTERNATING FILLS (UpDown or DownUp)")
print("="*70)

trades_sorted = trades.sort_values('unix_timestamp').reset_index(drop=True)
trades_sorted['prev_outcome'] = trades_sorted['outcome'].shift(1)
trades_sorted['time_since_last'] = trades_sorted['unix_timestamp'].diff()
trades_sorted['is_alternating'] = trades_sorted['outcome'] != trades_sorted['prev_outcome']

alternating = trades_sorted[trades_sorted['is_alternating'] == True]
same_side = trades_sorted[trades_sorted['is_alternating'] == False]

print(f"\nAlternating fills (switched sides): {len(alternating)}")
print(f"Same-side consecutive fills: {len(same_side)}")

print("\n=== Time Gap When ALTERNATING (UpDown or DownUp) ===")
print(alternating['time_since_last'].describe())

print("\n=== Time Gap When SAME SIDE (UpUp or DownDown) ===")
print(same_side['time_since_last'].describe())

# Burst detection - multiple fills in quick succession
print("\n" + "="*70)
print("BURST DETECTION")
print("="*70)

trades_sorted['burst_group'] = (trades_sorted['time_since_last'] > 2).cumsum()
bursts = trades_sorted.groupby('burst_group').agg({
    'size': ['count', 'sum'],
    'unix_timestamp': ['min', 'max'],
    'outcome': lambda x: x.value_counts().to_dict()
}).reset_index()
bursts.columns = ['burst_id', 'num_fills', 'total_size', 'start_time', 'end_time', 'outcome_breakdown']
bursts['duration'] = bursts['end_time'] - bursts['start_time']

print(f"\nTotal bursts detected (>2s gap between): {len(bursts)}")
print(f"\n=== Fills Per Burst ===")
print(bursts['num_fills'].describe())

print(f"\n=== Burst Duration (seconds) ===")
print(bursts['duration'].describe())

large_bursts = bursts[bursts['num_fills'] >= 5]
print(f"\n=== Large Bursts (5+ fills) ===")
print(f"Count: {len(large_bursts)}")
print(large_bursts[['num_fills', 'total_size', 'duration', 'outcome_breakdown']].head(10))

print("\n" + "="*70)
print("INTERPRETATION:")
print("="*70)
print("""
- Short same-side gaps (<1s): Orders being refilled immediately after execution
- Longer gaps (>5s): Waiting for market or order book to refresh
- Burst patterns: Multiple orders queued and filled against incoming market order
- Alternating patterns: Both sides being maintained simultaneously

Key insight: If same-side gaps are very short (<1s), they're likely maintaining 
a constant bid that gets refilled immediately when hit.
""")
```

```python
"""
TEST 3: ORDER TYPE INFERENCE
============================
Question: What order type are they using? GTC, FOK, FAK?

Method:
- GTC: Orders rest in book, get filled over time, partial fills common
- FOK: All-or-nothing, no partial fills at same timestamp
- FAK: Partial fills ok, but unfilled portion cancelled immediately

We can infer from:
- Partial fill frequency
- Whether same-size orders repeat (reposted GTC)
- Fill patterns at same timestamp
"""

import pandas as pd
import numpy as np

markets = ["1766688300", "1766689200", "1766690100", "1766691000", "1766691900", "1766692800"]

all_trades = []
for market in markets:
    try:
        trades = pd.read_parquet(f"DataCollection/BotTrades/{market}-trades.parquet")
        trades['market'] = market
        all_trades.append(trades)
    except:
        pass

trades = pd.concat(all_trades, ignore_index=True)

print("="*70)
print("TEST 3: ORDER TYPE INFERENCE")
print("="*70)

# Check if sizes are round numbers (likely full fills) or fractional (partial fills)
trades['is_round'] = (trades['size'] == trades['size'].round(0))
trades['is_max_size'] = (trades['size'] == 28.0)

print("\n=== SIZE PATTERNS ===")
print(f"Total trades: {len(trades)}")
print(f"Round number sizes: {trades['is_round'].sum()} ({trades['is_round'].mean()*100:.1f}%)")
print(f"Fractional sizes: {(~trades['is_round']).sum()} ({(~trades['is_round']).mean()*100:.1f}%)")
print(f"Exactly 28 shares: {trades['is_max_size'].sum()} ({trades['is_max_size'].mean()*100:.1f}%)")

print("\n=== FRACTIONAL SIZE ANALYSIS ===")
fractional = trades[~trades['is_round']]['size']
print(f"Count: {len(fractional)}")
print(fractional.describe())

# Most common fractional sizes
print("\nMost common fractional sizes:")
print(fractional.value_counts().head(15))

# Check for "remainder" patterns
# If they submit 28 and get partially filled, remainders should cluster
print("\n=== REMAINDER PATTERN CHECK ===")
# Common remainders if 28 is split
fractional_dist_from_28 = 28.0 - fractional
print("Distance from 28 for fractional fills:")
print(fractional_dist_from_28.describe())

# Group by timestamp to see if multiple fills happen together
print("\n=== FILLS PER TIMESTAMP ===")
by_ts = trades.groupby(['market', 'unix_timestamp']).size()
print(by_ts.describe())
print("\nDistribution:")
print(by_ts.value_counts().sort_index().head(15))

# Check for partial fill patterns at same timestamp
print("\n=== SAME-TIMESTAMP FILL ANALYSIS ===")
multi_fill_ts = trades.groupby(['market', 'unix_timestamp']).filter(lambda x: len(x) > 1)
print(f"Trades that occurred with other fills at same timestamp: {len(multi_fill_ts)}")

if len(multi_fill_ts) > 0:
    # Are these the same side or different sides?
    same_ts_groups = multi_fill_ts.groupby(['market', 'unix_timestamp'])
    
    same_side_count = 0
    diff_side_count = 0
    mixed_count = 0
    
    for _, group in same_ts_groups:
        outcomes = group['outcome'].unique()
        if len(outcomes) == 1:
            same_side_count += 1
        else:
            mixed_count += 1
    
    print(f"\nSame-timestamp groups - same side only: {same_side_count}")
    print(f"Same-timestamp groups - both sides: {mixed_count}")

# FOK vs FAK detection
print("\n=== FOK vs FAK DETECTION ===")
# FOK: If we see partial fills, it's NOT FOK
# FAK: Partial fills allowed, would see various sizes

# Look at size distribution
print("\nSize distribution (if mostly 28s with rare partials  GTC getting partially hit)")
print("(if wide variety of sizes  FAK or market orders hitting various depths)")
print(trades['size'].value_counts().head(20))

# Check if they ever get the same size twice in a row quickly (GTC refill pattern)
trades_sorted = trades.sort_values('unix_timestamp')
trades_sorted['prev_size'] = trades_sorted['size'].shift(1)
trades_sorted['time_gap'] = trades_sorted['unix_timestamp'].diff()
trades_sorted['same_size_as_prev'] = (trades_sorted['size'] == trades_sorted['prev_size'])

quick_repeats = trades_sorted[(trades_sorted['time_gap'] < 5) & (trades_sorted['same_size_as_prev'])]
print(f"\n=== QUICK REPEAT PATTERN (same size within 5s) ===")
print(f"Count: {len(quick_repeats)}")
print(f"Most common repeated sizes:")
print(quick_repeats['size'].value_counts().head(10))

print("\n" + "="*70)
print("INTERPRETATION:")
print("="*70)
print("""
ORDER TYPE SIGNATURES:

GTC (Good-Til-Cancelled):
- Orders rest in book, filled when someone crosses
- Partial fills common (someone takes only part of your order)
- Same sizes repeat as orders are refilled
-  Evidence: We see lots of 28-share repeats, some partials

FOK (Fill-Or-Kill):
- All or nothing - no partial fills
- Would see ONLY round numbers matching their intended size
-  Evidence: We see fractional sizes, so NOT FOK

FAK (Fill-And-Kill / IOC):
- Partial fills ok, but unfilled portion cancelled
- Used for aggressive orders crossing the spread
- Would see variety of partial fill sizes
- ? Evidence: Fractional sizes exist, but could be from book depth

LIKELY ANSWER: GTC limit orders
- 28-share orders posted at bid
- Get filled fully (28) or partially (fractional) when someone sells
- Reposted after fills
""")
```

```python
"""
TEST 5: STALE QUOTE DETECTION & CANCEL BEHAVIOR
================================================
Question: When do they cancel stale orders? What triggers a requote?

Method:
- Track fills relative to orderbook changes
- Identify when fills happen at "old" prices
- Infer cancel threshold from price gaps
"""

import pandas as pd
import numpy as np

markets = ["1766688300", "1766689200", "1766690100", "1766691000", "1766691900", "1766692800"]

all_data = []

for market in markets:
    try:
        trades = pd.read_parquet(f"DataCollection/BotTrades/{market}-trades.parquet")
        orderbook = pd.read_parquet(f"DataCollection/OrderbookData/{market}-orderbook.parquet")
        
        trades['market'] = market
        trades = trades.sort_values('unix_timestamp')
        
        orderbook = orderbook.sort_values('unixtime')
        orderbook['unix_ts'] = orderbook['unixtime'].astype(int)
        
        # Calculate orderbook changes
        orderbook['up_bid_change'] = orderbook['up_best_bid'].diff()
        orderbook['down_bid_change'] = orderbook['down_best_bid'].diff()
        orderbook['up_bid_pct_change'] = orderbook['up_best_bid'].pct_change()
        orderbook['down_bid_pct_change'] = orderbook['down_best_bid'].pct_change()
        
        # Merge trades with orderbook state
        merged = pd.merge_asof(
            trades,
            orderbook[['unix_ts', 'up_best_bid', 'up_best_ask', 'down_best_bid', 'down_best_ask',
                      'up_bid_change', 'down_bid_change', 'up_bid_pct_change', 'down_bid_pct_change']],
            left_on='unix_timestamp',
            right_on='unix_ts',
            direction='backward'
        )
        
        all_data.append(merged)
    except Exception as e:
        print(f"Error processing {market}: {e}")

df = pd.concat(all_data, ignore_index=True)

print("="*70)
print("TEST 5: STALE QUOTE ANALYSIS")
print("="*70)

# For each trade, check how far the fill price is from current best bid
df['current_bid'] = np.where(df['outcome'] == 'Up', df['up_best_bid'], df['down_best_bid'])
df['price_vs_current_bid'] = df['price'] - df['current_bid']

print("\n=== FILL PRICE vs CURRENT BEST BID ===")
print(df['price_vs_current_bid'].describe())

print("\n=== DISTRIBUTION ===")
conditions = [
    (df['price_vs_current_bid'] < -0.02, 'Filled 2%+ below current bid (got better price)'),
    (df['price_vs_current_bid'] < -0.01, 'Filled 1-2% below current bid'),
    (df['price_vs_current_bid'].abs() < 0.01, 'Filled near current bid (1%)'),
    (df['price_vs_current_bid'] > 0.01, 'Filled 1-2% above current bid'),
    (df['price_vs_current_bid'] > 0.02, 'Filled 2%+ above current bid (stale order?)'),
]

for condition, label in conditions:
    count = condition.sum()
    print(f"{label}: {count} ({count/len(df)*100:.1f}%)")

# Look at fills that happened when market had just moved
print("\n=== FILLS AFTER LARGE ORDERBOOK MOVES ===")

# Recent bid change before fill
df['recent_up_change'] = df['up_bid_pct_change'].abs()
df['recent_down_change'] = df['down_bid_pct_change'].abs()

# High change = market just moved
high_change_threshold = df['recent_up_change'].quantile(0.9)

print(f"\nThreshold for 'large move': {high_change_threshold:.4f}")

for outcome in ['Up', 'Down']:
    subset = df[df['outcome'] == outcome]
    if outcome == 'Up':
        after_big_move = subset[subset['recent_up_change'] > high_change_threshold]
        normal = subset[subset['recent_up_change'] <= high_change_threshold]
    else:
        after_big_move = subset[subset['recent_down_change'] > high_change_threshold]
        normal = subset[subset['recent_down_change'] <= high_change_threshold]
    
    print(f"\n{outcome} trades:")
    print(f"  Normal conditions: {len(normal)} trades, avg price_vs_bid: {normal['price_vs_current_bid'].mean():.4f}")
    print(f"  After big move: {len(after_big_move)} trades, avg price_vs_bid: {after_big_move['price_vs_current_bid'].mean():.4f}")

# Check for patterns suggesting they cancel after X% move
print("\n=== INFERRING CANCEL THRESHOLD ===")

# If they cancel stale orders, we should see fewer fills at prices far from current bid
# Let's bin by how far the fill was from current bid and see the distribution

df['staleness'] = (df['price'] - df['current_bid']).abs()
bins = [0, 0.005, 0.01, 0.02, 0.03, 0.05, 0.1, 1]
labels = ['<0.5%', '0.5-1%', '1-2%', '2-3%', '3-5%', '5-10%', '>10%']
df['staleness_bucket'] = pd.cut(df['staleness'], bins=bins, labels=labels)

print("\nFill staleness distribution:")
print(df['staleness_bucket'].value_counts().sort_index())

# Time since last fill on same side when fill is "stale"
print("\n=== TIME PATTERN FOR STALE FILLS ===")
stale_fills = df[df['staleness'] > 0.02]  # Fills more than 2% from current bid
fresh_fills = df[df['staleness'] <= 0.02]

print(f"Fresh fills (<2% from bid): {len(fresh_fills)}")
print(f"Stale fills (>2% from bid): {len(stale_fills)}")

if len(stale_fills) > 0:
    print(f"\nStale fills - what outcome?")
    print(stale_fills['outcome'].value_counts())

print("\n" + "="*70)
print("INTERPRETATION:")
print("="*70)
print("""
STALE QUOTE BEHAVIOR:

1. If most fills are near current bid (1%):
    They're updating quotes quickly, or market is stable

2. If we see fills 2%+ from current bid:
    Either they leave orders resting (GTC)
    Or our orderbook snapshots are lagged

3. If stale fills cluster on one side:
    They're getting picked off (adverse selection)

4. The "staleness distribution" suggests their quote life:
   - Sharp dropoff at X%  they cancel at X% move
   - Flat distribution  they let orders rest indefinitely

KEY INSIGHT: The cancel threshold is a critical parameter.
Too tight  miss fills. Too loose  get picked off.
""")
```
